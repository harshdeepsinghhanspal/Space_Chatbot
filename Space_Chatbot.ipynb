```python
def handle_input(user_input):
    try:
        if not any(keyword in user_input.lower() for keyword in space_keywords):
            return "🚀 Please ask about space-related topics only! (Agencies, planets, rockets, satellites, space DIY etc.)"

        # If it's about space, invoke the model
        ai_response = chain.invoke({"input": user_input})
        return ai_response

    except Exception as e:
        return f"❌ Error: {str(e)}"
```

```python
print(handle_input("please guide me what prompts to ask?"))
```

```python
def chat_interface(user_msg, history):
    try:
        reply = handle_input(user_msg)
        return reply  # ONLY return the reply string
    except Exception as e:
        import traceback
        print("⚠️ Error in chat_interface:", traceback.format_exc())
        return f"❌ Error: {str(e)}"
```

```python
gr.ChatInterface(
    fn=chat_interface,
    title="🚀 SpaceBot - Ask Me Anything About Space!"
).launch(share=True, debug=True)
```
